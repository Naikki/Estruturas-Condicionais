
### Ímpar ou par
Esse código permite receber um número inteiro, e mostra se ele é par ou ímpar.

```python
numero_str = input('Digite o número: ')
try:
    numero = int(numero_str)
    if numero % 2 == 0:
        print('Esse número é par')
    else:
        print('Esse número é ímpar')
except ValueError:
    print('Entrada inválida. Por favor, digite um número inteiro.')
```


### Checagem de acesso
Esse código permite que o usuário digite uma senha, e em seguida mostra se a senha digitada tem acesso permitido ou negado pelo sistema.

```python
senha_digitada_str = input('Digite a senha: ')
try:
    senha_digitada = int(senha_digitada_str)
    match senha_digitada:
        case 1234:
            print('Acesso permitido')
        case _:
            print('Acesso negado')
except ValueError:
    print('Entrada inválida.')
```


### Cálculo de média semestral
Código que recebe as duas notas bimestrais, e em seguida mostra a condição, caso tenha sido aprovado ou reprovado, caso..:
* **a média for >= 7**

```python
nota1 = float(input("Digite a primeira média bimestral: "))
nota2 = float(input("Digite a segunda média bimestral: "))

# Cálculo da média semestral
media = (nota1 + nota2) / 2

# Exibição detalhada dos resultados
print("\n" + "="*39)
print("RESULTADO FINAL:")
print("="*39)
print(f"1º Bimestre: {nota1:.1f}")
print(f"2º Bimestre: {nota2:.1f}")
print(f"Média semestral: {media:.1f}")
print("-"*39)

# Verificação da situação do aluno
if media >= 7:
    print(f"Situação: APROVADO com média: {media:.1f}")
else:
    print(f"Situação: REPROVADO com média: {media:.1f}")
print("="*39)
```

### Cálculo de Peso Ideal

Este código solicita a altura (em metros) e o sexo de uma pessoa e calcula o peso ideal utilizando as seguintes fórmulas:

*   **Para homens:** `(72.7 * altura) - 58`
*   **Para mulheres:** `(62.1 * altura) - 44.7`

```python
try:
    altura_str = input('Digite a sua altura (ex: 1.75): ')
    altura = float(altura_str)

    if altura <= 0:
        print('A altura deve ser um valor positivo.')
    else:
        sexo = input('Digite o sexo (M para masculino, F para feminino): ').upper()

        if sexo == 'M':
            peso_ideal = (72.7 * altura) - 58
            print(f'Para um homem com {altura:.2f}m de altura, o peso ideal é: {peso_ideal:.2f} kg')
        elif sexo == 'F':
            peso_ideal = (62.1 * altura) - 44.7
            print(f'Para uma mulher com {altura:.2f}m de altura, o peso ideal é: {peso_ideal:.2f} kg')
        else:
            print('Sexo inválido. Por favor, digite M para masculino ou F para feminino.')

except ValueError:
    print('Entrada inválida. Certifique-se de digitar os dados corretamente.')
except Exception as e:
    print(f'Ocorreu um erro inesperado: {e}')
```

### Ordem Decrescente de Três Números Distintos
Este código lê três números inteiros e distintos fornecidos pelo usuário e os exibe em ordem decrescente, utilizando uma estrutura de seleção encadeada (if/elif/else).

```python
try:
    num1_str = input('Digite o primeiro número inteiro: ')
    num1 = int(num1_str)

    num2_str = input('Digite o segundo número inteiro (diferente do primeiro): ')
    num2 = int(num2_str)
    while num2 == num1:
        print('O segundo número deve ser diferente do primeiro. Por favor, digite novamente.')
        num2_str = input('Digite o segundo número inteiro (diferente do primeiro): ')
        num2 = int(num2_str)

    num3_str = input('Digite o terceiro número inteiro (diferente dos anteriores): ')
    num3 = int(num3_str)
    while num3 == num1 or num3 == num2:
        print('O terceiro número deve ser diferente dos anteriores. Por favor, digite novamente.')
        num3_str = input('Digite o terceiro número inteiro (diferente dos anteriores): ')
        num3 = int(num3_str)

    print(f'\nOs números digitados são: {num1}, {num2}, {num3}')
    print('Ordenando em ordem decrescente...')

    # Lógica para ordenar em ordem decrescente usando seleção encadeada
    if num1 >= num2 and num1 >= num3:
        if num2 >= num3:
            print(f'Ordem decrescente: {num1}, {num2}, {num3}')
        else:
            print(f'Ordem decrescente: {num1}, {num3}, {num2}')
    elif num2 >= num1 and num2 >= num3:
        if num1 >= num3:
            print(f'Ordem decrescente: {num2}, {num1}, {num3}')
        else:
            print(f'Ordem decrescente: {num2}, {num3}, {num1}')
    else:
        # num3 é o maior
        if num1 >= num2:
            print(f'Ordem decrescente: {num3}, {num1}, {num2}')
        else:
            print(f'Ordem decrescente: {num3}, {num2}, {num1}')

except ValueError:
    print('Entrada inválida. Por favor, digite apenas números inteiros.')
except Exception as e:
    print(f'Ocorreu um erro inesperado: {e}')
```

### Verificação de Elegibilidade para Campeonato de Bolinha de Gude (Categoria Ouro)

Este código verifica se um jogador está apto a participar da "categoria ouro" do primeiro campeonato mundial de bolinha de gude, com base em seu peso e altura. As condições são:

*   **Peso**: entre 70 kg e 80 kg (inclusive).
*   **Altura**: entre 1,75 m e 1,90 m (inclusive).

```python
try:
    peso_str = input('Digite o peso do jogador em kg (ex: 75.5): ')
    peso = float(peso_str)

    altura_str = input('Digite a altura do jogador em metros (ex: 1.80): ')
    altura = float(altura_str)

    peso_valido = False
    altura_valida = False

    # Verifica o peso
    if 70 <= peso <= 80:
        peso_valido = True

    # Verifica a altura
    if 1.75 <= altura <= 1.90:
        altura_valida = True

    # Determina a mensagem com base na validade do peso e altura
    if peso_valido and altura_valida:
        print('Aceito')
    elif not peso_valido and altura_valida:
        print('Recusado por peso')
    elif peso_valido and not altura_valida:
        print('Recusado por altura')
    else:
        print('Recusado por peso e altura')

except ValueError:
    print('Entrada inválida. Por favor, digite valores numéricos válidos para peso e altura.')
except Exception as e:
    print(f'Ocorreu um erro inesperado: {e}')
```

### Calculadora dos Preços das Frutas com Desconto

Este código calcula o valor total a ser pago por um cliente que compra morangos e maçãs, aplicando diferentes preços por quilo com base na quantidade e um desconto de 20% se a compra atender a certas condições.

```python
try:
    # Recebe a quantidade de morangos
    morango_kg_str = input('Digite a quantidade de morangos em kg: ')
    morango_kg = float(morango_kg_str)

    # Recebe a quantidade de maçãs
    maca_kg_str = input('Digite a quantidade de maçãs em kg: ')
    maca_kg = float(maca_kg_str)

    # Inicializa os preços
    preco_morango = 0
    preco_maca = 0

    # Calcula o preço dos morangos
    if morango_kg <= 5:
        preco_morango = morango_kg * 5.00 # R$ 5,00 por kg
    else:
        preco_morango = morango_kg * 4.00 # R$ 4,00 por kg

    # Calcula o preço das maçãs
    if maca_kg <= 5:
        preco_maca = maca_kg * 3.00 # R$ 3,00 por kg
    else:
        preco_maca = maca_kg * 2.00 # R$ 2,00 por kg

    # Calcula o valor total antes do desconto
    valor_total_sem_desconto = preco_morango + preco_maca

    # Calcula a quantidade total de frutas
    quantidade_total_kg = morango_kg + maca_kg

    # Verifica as condições para o desconto
    desconto = 0
    if quantidade_total_kg > 8 or valor_total_sem_desconto > 35:
        desconto = valor_total_sem_desconto * 0.20 # 20% de desconto

    # Calcula o valor final a ser pago
    valor_final_a_pagar = valor_total_sem_desconto - desconto

    print(f'\n--- Detalhes da Compra ---')
    print(f'Morangos: {morango_kg:.0f} kg')
    print(f'Maçãs: {maca_kg:.0f} kg')
    print(f'Preço Morangos: R$ {preco_morango:.2f}')
    print(f'Preço Maçãs: R$ {preco_maca:.2f}')
    print(f'Valor Total (sem desconto): R$ {valor_total_sem_desconto:.2f}')
    if desconto > 0:
        print(f'Desconto Aplicado: R$ {desconto:.2f}')
    print(f'-------------------------')
    print(f'Valor a Pagar: R$ {valor_final_a_pagar:.2f}')

except ValueError:
    print('Entrada inválida. Por favor, digite quantidades numéricas válidas.')
except Exception as e:
    print(f'Ocorreu um erro inesperado: {e}')
```

### Organização de Três Números com Base em um Critério

Este código solicita quatro valores: um inteiro `I` (que define a regra de ordenação: 1 para crescente, 2 para decrescente, 3 para o maior no meio) e três números reais `A`, `B`, `C`. Ele então imprime `A`, `B` e `C` de acordo com a regra selecionada por `I`.

```python
try:
    # Entrada dos valores
    i_str = input('Digite o valor de I (1, 2 ou 3): ')
    I = int(i_str)

    a_str = input('Digite o valor de A: ')
    A = float(a_str)

    b_str = input('Digite o valor de B: ')
    B = float(b_str)

    c_str = input('Digite o valor de C: ')
    C = float(c_str)

    numeros = [A, B, C]

    print(f'\nValores originais: A={A}, B={B}, C={C}')

    # Lógica para organizar e imprimir os números
    if I == 1:
        # Ordem crescente
        numeros.sort()
        print(f'Ordem crescente: {numeros[0]}, {numeros[1]}, {numeros[2]}')
    elif I == 2:
        # Ordem decrescente
        numeros.sort(reverse=True)
        print(f'Ordem decrescente: {numeros[0]}, {numeros[1]}, {numeros[2]}')
    elif I == 3:
        # O maior fica entre os outros dois
        max_val = max(numeros)
        min_val = min(numeros)
        # Encontra o valor do meio (o que não é nem o máximo nem o mínimo)
        mid_val = sum(numeros) - max_val - min_val
        print(f'Maior entre os outros: {min_val}, {max_val}, {mid_val}')
    else:
        print('Valor de I inválido. Por favor, digite 1, 2 ou 3.')

except ValueError:
    print('Entrada inválida. Por favor, digite um número inteiro para I e números reais para A, B e C.')
except Exception as e:
    print(f'Ocorreu um erro inesperado: {e}')
```

# Programa de Cálculos Salariais

Este código implementa um sistema de menu interativo para realizar diferentes operações relacionadas a salários de funcionários. A estrutura do programa segue os seguintes passos:

1. Exibição do Menu: O programa apresenta um menu com 3 opções para o usuário, cada uma realizando uma operação diferente com o salário informado.
2. Entrada da Opção: O usuário digita a opção desejada (1, 2 ou 3), e o programa valida se a opção é válida.
3. Tratamento das Opções:
   · Opção 1: Calcula o imposto a ser pago baseado em faixas salariais e mostra o salário final após o desconto.
   · Opção 2: Calcula o novo salário com aumento baseado em diferentes faixas salariais.
   · Opção 3: Classifica o salário como "Mal remunerado" ou "Bem remunerado".
4. Controle de Fluxo: O programa utiliza estruturas condicionais (if-elif-else) para direcionar a execução conforme a opção escolhida e as faixas salariais.
5. Tratamento de Erros: Verifica se a opção digitada é válida (1, 2 ou 3) e exibe uma mensagem de erro caso contrário.
6. Repetição: O programa fica em loop até que o usuário decida sair, permitindo realizar múltiplas operações sem reiniciar

```python

def formatar_moeda(valor):
    """Formata um valor numérico como moeda brasileira (R$)"""
    return f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

def converter_para_float(valor_str):
    """Converte uma string no formato brasileiro para float"""
    # Remove o símbolo R$ se existir
    valor_str = valor_str.replace('R$', '').strip()

    # Remove pontos (separadores de milhar)
    valor_str = valor_str.replace('.', '')

    # Substitui vírgula (separador decimal) por ponto
    valor_str = valor_str.replace(',', '.')

    # Converte para float
    try:
        return float(valor_str)
    except ValueError:
        raise ValueError("Valor inválido! Digite o salário no formato correto (ex: 1.450,00 ou 1450,00)")

def main():
    while True:
        # Exibe o menu de opções
        print("\n" + "="*50)
        print("MENU DE OPÇÕES".center(50))
        print("="*50)
        print("1 - Imposto a ser pago e salário final")
        print("2 - Novo salário com aumento")
        print("3 - Classificação")
        print("4 - Sair do programa")
        print("="*50)

        # Recebe a opção do usuário
        try:
            opcao = int(input("Digite a opção desejada: "))
        except ValueError:
            print("Opção inválida! Digite um número entre 1 e 4.")
            continue

        # Opção para sair do programa
        if opcao == 4:
            print("Programa encerrado. Até logo!")
            break

        # Verifica se a opção é válida
        if opcao not in [1, 2, 3, 4]:
            print("Opção inválida! Digite um número entre 1 e 4.")
            continue

        # Se a opção for 1, 2 ou 3, solicita o salário
        if opcao in [1, 2, 3]:
            try:
                salario_str = input("Digite o salário do funcionário (ex: 1.450,00): R$ ")
                salario = converter_para_float(salario_str)
            except ValueError as e:
                print(str(e))
                continue

        # Opção 1: Cálculo do imposto
        if opcao == 1:
            print("\n" + "-"*50)
            print("CÁLCULO DE IMPOSTO E SALÁRIO FINAL".center(50))
            print("-"*50)

            # Calcula o imposto conforme as faixas
            if salario < 1100:
                imposto = salario * 0.05
                taxa_imposto = "5%"
            elif salario <= 3000:
                imposto = salario * 0.10
                taxa_imposto = "10%"
            else:
                imposto = salario * 0.15
                taxa_imposto = "15%"

            salario_final = salario - imposto

            # Exibe os resultados formatados
            print(f"Salário informado: {formatar_moeda(salario)}")
            print(f"Taxa de imposto aplicada: {taxa_imposto}")
            print(f"Imposto a pagar: {formatar_moeda(imposto)}")
            print(f"Salário final: {formatar_moeda(salario_final)}")

        # Opção 2: Cálculo do novo salário com aumento
        elif opcao == 2:
            print("\n" + "-"*50)
            print("CÁLCULO DO NOVO SALÁRIO COM AUMENTO".center(50))
            print("-"*50)

            # Calcula o aumento conforme as faixas
            if salario > 3000:
                aumento = 450
            elif salario >= 2000:
                aumento = 300
            elif salario >= 1500:
                aumento = 250
            else:
                aumento = 100

            novo_salario = salario + aumento

            # Exibe os resultados formatados
            print(f"Salário atual: {formatar_moeda(salario)}")
            print(f"Valor do aumento: {formatar_moeda(aumento)}")
            print(f"Novo salário: {formatar_moeda(novo_salario)}")

        # Opção 3: Classificação do salário
        elif opcao == 3:
            print("\n" + "-"*50)
            print("CLASSIFICAÇÃO SALARIAL".center(50))
            print("-"*50)

            # Classifica o salário
            if salario <= 1500:
                classificacao = "Mal remunerado"
            else:
                classificacao = "Bem remunerado"

            # Exibe os resultados formatados
            print(f"Salário informado: {formatar_moeda(salario)}")
            print(f"Classificação: {classificacao}")

if __name__ == "__main__":
    main()
```
